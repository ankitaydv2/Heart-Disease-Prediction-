```python
df = pd.read_csv('/kaggle/input/datasets/ankitayadav2005/heart-disease-prediction/heart.csv')
```

```python
df.head()
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(df.iloc[:, 0:-1],df.iloc[:,-1],test_size=0.2,random_state =2)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
```

```python
c1f1 = LogisticRegression(max_iter=1000)
c1f2 = DecisionTreeClassifier()
```

```python
DecisionTreeClassifier()
```

```python
c1f1.fit(X_train, y_train)
c1f2.fit(X_train, y_train )
```

```python
y_pred1 = c1f1.predict(X_test)
y_pred2 = c1f2.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score, confusion_matrix
print("Accuracy of Logistic Regression", accuracy_score(y_test, y_pred1))
print("Accuracy of Decision Trees", accuracy_score(y_test, y_pred2))
```
